## ¿Qué estado tiene el más alto número de indigentes por cada 10000 habitantes?

```python
#Número de indigentes por cada 10,000 habitantes
homelessness["indiv_per_10k"] = 10000 * homelessness['individuals'] / homelessness['state_pop'] 
#Que el número de indigentes sea mayor que 20
high_homelessness = homelessness[homelessness['indiv_per_10k'] > 20]
#Ordenar de forma descendente
high_homelessness_srt = high_homelessness.sort_values('indiv_per_10k', ascending = False)
#Seleccionar columnas "state" e "indiv_per_10k"
result = high_homelessness_srt[['state', 'indiv_per_10k']]
print(result)
```

Washington, D.C. tiene el número más grande de indigentes con casi 54 personas por cada 10000 habitantes.

```python
sales = pd.read_csv('sales_subset.csv', index_col=0)
```

```python
print(sales.head())
print(sales.info())
print(sales['weekly_sales'].mean())
print(sales['weekly_sales'].median())
```

¡La cantidad media de ventas semanales es casi el doble de la cantidad media de ventas semanales! Esto puede indicarle que hay algunas semanas de ventas muy altas que hacen que la media sea mucho más alta que la mediana.

```python
#Columnas del dataframe
sales.columns
```

```python
#Los límites superior e inferior de la fecha
print(sales['date'].max())
print(sales['date'].min())
```

```python
#Rango interquartílico
def iqr(column):
    return column.quantile(0.75) - column.quantile(0.25)
```

```python
#IQR de la temperatura, precio de la gasolina por litro y tasa de desempleo
print(sales['temperature_c'].agg(iqr))
print(sales[["temperature_c", 'fuel_price_usd_per_l', 'unemployment']].agg(iqr))
```

```python
#Lo mismo que lo anterior, pero también con la mediana
import numpy as np
print(sales[["temperature_c", "fuel_price_usd_per_l", "unemployment"]].agg([iqr, np.median]))
```

```python
#Calculará la suma acumulada y el máximo acumulativo de las ventas semanales de un departamento.
sales_1_1 = sales.sort_values('date')
sales_1_1['cum_weekly_sales'] = sales_1_1['weekly_sales'].cumsum()
sales_1_1['cum_max_sales'] = sales_1_1['weekly_sales'].cummax()
print(sales_1_1[["date", "weekly_sales", "cum_weekly_sales", "cum_max_sales"]])
```

```python
sales.head()
```

```python
#Eliminemos los duplicados de 'store' y 'type'
store_types = sales.drop_duplicates(['store', 'type'])
print(store_types.head())
#Eliminar duplicados de 'store' y 'department'
store_depts = sales.drop_duplicates(['store', 'department'])
print(store_depts.head(),'\n')
#Las fechas que son vacacionales y eliminar los duplicados
holiday_dates = sales[sales['is_holiday'] == True].drop_duplicates('date')
print(holiday_dates['date'])
```

```python
#Contar el número de tiendas en cada tipo
store_counts = store_types['type'].value_counts()
print(store_counts)
#Obtener la proporción de tiendas por tipo
store_props = store_types['type'].value_counts(normalize = True)
print(store_props)
#Contar el número de diferentes departamentos en "store_depts" y ordenados
dept_counts_sorted = store_depts['department'].value_counts(sort = True)
print(dept_counts_sorted)
#Obtener la proporción de departamentos de cada 'store' y ordenados
dept_props_sorted = store_depts['department'].value_counts(sort=True, normalize=True)
print(dept_props_sorted)
```

Parece ser que el departamento 43 existe únicamente en dos tiendas.

## ¿Qué porcentaje de ventas ocurrió en cada tipo de tienda en Walmart?

```python
#Forma no óptima

#Calculemos el total
sales_all = sales["weekly_sales"].sum()
#El total por cada tipo
sales_A = sales[sales["type"] == "A"]["weekly_sales"].sum()
sales_B = sales[sales["type"] == "B"]["weekly_sales"].sum()
sales_C = sales[sales["type"] == "C"]["weekly_sales"].sum()
#Obtengamos las proporciones
sales_propn_by_type = [sales_A, sales_B, sales_C] / sales_all
print(sales_propn_by_type)
```

```python
#Forma óptima
sales_by_type = sales.groupby('type')['weekly_sales'].sum()
sales_prop = sales_by_type/sum(sales['weekly_sales'])
print(sales_prop)
```

```python
#Agrupar por 'type' e 'is_holiday'
sales_by_type_is_holiday = sales.groupby(['type','is_holiday'])['weekly_sales'].sum()
print(sales_by_type_is_holiday)
```

```python
#Para cada tipo de tienda queremos el máx, min y la media de las ventas
sales_stats = sales.groupby('type')['weekly_sales'].agg([np.min, np.max, np.mean, np.median])
print(sales_stats)
#Para cada tipo de tienda queremos el máx, min y la media de 'unemployment' y 'fuel_price_usd_per_l'
unemp_fuel_stats = sales.groupby('type')[['unemployment','fuel_price_usd_per_l']].agg([np.min, np.max, np.mean, np.median])
print(unemp_fuel_stats)
```

```python
#Calcular la media de 'weekly_sales' por 'type' usando pivot_table
mean_sales_by_type = sales.pivot_table(values = 'weekly_sales', index = 'type')
print(mean_sales_by_type)
```

```python
#Lo mismo que lo anterior pero también con la mediana
mean_med_sales_by_type = sales.pivot_table(values = 'weekly_sales', index = 'type', aggfunc = [np.mean, np.median])
print(mean_med_sales_by_type)
```

```python
#Media de las ventas por 'type' y 'is_holiday'
mean_sales_by_type_holiday = sales.pivot_table(values = 'weekly_sales', index = 'type', columns = 'is_holiday')
print(mean_sales_by_type_holiday)
```

```python
#Media de 'weekly_sales' por 'department' y 'type' y llenando los valores faltantes con 0
print(sales.pivot_table(values = 'weekly_sales', index = 'department', columns = 'type', fill_value = 0))
```

```python
#Media de 'weekly_sales' por 'department' y type, llenando los faltantes con 0 y con los totales de columns y filaes
print(sales.pivot_table(values="weekly_sales", index="department", columns="type", fill_value = 0, margins = True))
```

```python
#Promedio de temperaturas por país, ciudad y año
temperatures = pd.read_csv('temperatures.csv', parse_dates = ['date'] , index_col = 0)
temperatures['year'] = temperatures['date'].dt.year
temp_by_country_city_vs_year = temperatures.pivot_table(values = 'avg_temp_c', index = ['country', 'city'], columns = 'year')
temp_by_country_city_vs_year
```

```python
#Seleccionar de Egypt a India
print(temp_by_country_city_vs_year.loc['Egypt':'India'])
#Seleccionar de Egypt, Cairo a India,Delhi
print(temp_by_country_city_vs_year.loc[('Egypt','Cairo'):('India','Delhi')])
# Seleccionar de Egypt,Cairo a India,Delhi de 2005 a 2010
temp_by_country_city_vs_year.loc[('Egypt','Cairo'):('India','Delhi'),'2005':'2010']
```

## El año con la temperatura media más alta y la ciudad con la temperatura media más baja

```python
#Temperatura media por año
mean_temp_by_year = temp_by_country_city_vs_year.mean()
#Año con la temperatura media más alta:
print(mean_temp_by_year[mean_temp_by_year == mean_temp_by_year.max()])
#Temperatura media por ciudad
mean_temp_by_city = temp_by_country_city_vs_year.mean(axis = 'columns')
#Ciudad con la temperatura más baja
print(mean_temp_by_city[mean_temp_by_city == mean_temp_by_city.min()])
```
